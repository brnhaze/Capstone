```python
import matplotlib.pyplot as plt

# Group by penalty name and sum penalty yards
penalty_yards_by_name = pos_2.groupby('penalty_name')['penalty_yards'].sum()

# Sort penalty yards in descending order
penalty_yards_by_name_sorted = penalty_yards_by_name.sort_values(ascending=False)

# Plot histogram
plt.figure(figsize=(10, 6))
bars = plt.bar(penalty_yards_by_name_sorted.index, penalty_yards_by_name_sorted.values, color='blue')
plt.xlabel('Penalty Name')
plt.ylabel('Total Penalty Yards')
plt.title('Total Penalty Yards per Penalty Name')

# Annotate each bar with its total penalty yards value
for bar, yards in zip(bars, penalty_yards_by_name_sorted.values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(yards), ha='center', va='bottom')

plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
